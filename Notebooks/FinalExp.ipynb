### Budget Sensitivity

#### Charlottesville City:

```python
dict_frame = {}

dict_frame["k"] = []
dict_frame["seed"] = []
dict_frame["algorithm"] = []
dict_frame["facilities"] = []
dict_frame["objective"] = []

cover_runs = [run for run in os.listdir(PROJECT_ROOT/"Experiments"/"output"/"budget_sensitivity"/"cover") if run.endswith('.json') and "cville" in run]
for file in cover_runs:
    with open(PROJECT_ROOT/"Experiments"/"output"/"budget_sensitivity"/"cover"/file) as f:
        data = json.load(f)
        seed = data[0]
        data = data[1]
        dict_frame["facilities"] += [val["paths"] for val in data.values()]
        dict_frame["seed"] += [seed for key in data.keys()]
        dict_frame["k"] += [key for key in data.keys()]
        dict_frame["algorithm"] += ["cover" for key in data.keys()]
        dict_frame["objective"] += [val["objective"] for val in data.values()]

supplier_runs = [run for run in os.listdir(PROJECT_ROOT/"Experiments"/"output"/"budget_sensitivity"/"supplier") if run.endswith('.json') and "cville" in run]
for file in supplier_runs:
    with open(PROJECT_ROOT/"Experiments"/"output"/"budget_sensitivity"/"supplier"/file) as f:
        data = json.load(f)
        seed = data[0]
        data = data[1]
        dict_frame["facilities"] += [val["paths"] for val in data.values()]
        dict_frame["seed"] += [seed for key in data.keys()]
        dict_frame["k"] += [key for key in data.keys()]
        dict_frame["algorithm"] += ["supplier" for key in data.keys()]
        dict_frame["objective"] += [val["objective"] for val in data.values()]
        
baseline_runs = [run for run in os.listdir(PROJECT_ROOT/"Experiments"/"output"/"budget_sensitivity"/"baseline") if run.endswith('.json') and "cville" in run]
for file in baseline_runs:
    with open(PROJECT_ROOT/"Experiments"/"output"/"budget_sensitivity"/"baseline"/file) as f:
        data = json.load(f)
        seed = data[0]
        data = data[1]
        dict_frame["facilities"] += [val["paths"] for val in data.values()]
        dict_frame["seed"] += [seed for key in data.keys()]
        dict_frame["k"] += [key for key in data.keys()]
        dict_frame["algorithm"] += ["baseline" for key in data.keys()]
        dict_frame["objective"] += [val["objective"] for val in data.values()]

flow_runs = [run for run in os.listdir(PROJECT_ROOT/"Experiments"/"output"/"budget_sensitivity"/"flow") if run.endswith('.json') and "cville" in run]
for file in flow_runs:
    with open(PROJECT_ROOT/"Experiments"/"output"/"budget_sensitivity"/"flow"/file) as f:
        data = json.load(f)
        seed = data[0]
        data = data[1]
        dict_frame["facilities"] += [val["paths"] for val in data.values()]
        dict_frame["seed"] += [seed for key in data.keys()]
        dict_frame["k"] += [key for key in data.keys()]
        dict_frame["algorithm"] += ["flow" for key in data.keys()]
        dict_frame["objective"] += [val["objective"] for val in data.values()]
```

### Percentile Coverage

#### Charlottesville City:

```python
dict_frame = {}
m = 5
k = 5
with open(PROJECT_ROOT/"Experiments"/"output"/f"percentile_sensitivity_{m}_{k}_narrow_robust_baseline.json") as f:
    data = json.load(f)
    percentile = data["percentile_list"]
    dict_frame["seed"] = [seed[0] for seed in data["cover"] for _ in percentile] + [seed[0] for seed in data["supplier"] for _ in percentile] + [seed[0] for seed in data["baseline"] for _ in percentile]  + [seed[0] for seed in data["flow"] for _ in percentile]
    dict_frame["algorithm"] = ["cover" for _ in data["cover"] for _ in percentile] + ["supplier" for _ in data["supplier"] for _ in percentile] + ["baseline" for _ in data["baseline"] for _ in percentile]  + ["flow" for _ in data["flow"] for _ in percentile]
    dict_frame["fac"] = [path[1]["paths"] for path in data["cover"] for _ in percentile] + [path[1]["paths"] for path in data["supplier"] for _ in percentile] + [path[1]["paths"] for path in data["baseline"] for _ in percentile]  + [path[1]["paths"] for path in data["flow"] for _ in percentile]
    dict_frame["percentile"] = [p for _ in data["cover"] for p in percentile] + [p for _ in data["supplier"] for p in percentile] + [p for _ in data["baseline"] for p in percentile]  + [p for _ in data["flow"] for p in percentile]
    dict_frame["objective"] = [obj for obj_list in data["cover"] for obj in obj_list[1]["obj_list"]] + [obj for obj_list in data["supplier"] for obj in obj_list[1]["obj_list"]] + [obj for obj_list in data["baseline"] for obj in obj_list[1]["obj_list"]] + [obj for obj_list in data["flow"] for obj in obj_list[1]["obj_list"]]
```

### Movement Sensitivity

##### Charlottesville City:

```python
dict_frame = {}

dict_frame["fac"] = []
dict_frame["objective"] = []
dict_frame["m"] = []
dict_frame["k"] = []
dict_frame["seed"] = []

runs = [run for run in os.listdir(PROJECT_ROOT/"Experiments"/"output"/"movement_sensitivity"/"cover") if run.endswith('.json') and "cville" in run]
for file in runs:
    with open(PROJECT_ROOT/"Experiments"/"output"/"movement_sensitivity"/"cover"/file) as f:
        data = json.load(f)
        seed = int(data["seed"])
        
        data = data["cover_runs"]
        m_values = data.keys()
        k_values = data[list(data.keys())[0]].keys()
        dict_frame["fac"] += [data[m][k]["paths"] for m in m_values for k in k_values]
        dict_frame["objective"] += [data[m][k]["objective"] for m in m_values for k in k_values]
        dict_frame["m"] += [float(m) for m in m_values for k in k_values]
        dict_frame["k"] += [int(k) for m in m_values for k in k_values]
        dict_frame["seed"] += [seed for m in m_values for k in k_values]
```
